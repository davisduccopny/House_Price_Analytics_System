```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

file_path = './TinhHuyenXa2021.xlsx'
df = pd.read_excel(file_path)
```

```python
columns_to_drop = ['Tên Tiếng Anh','Cấp']

df = df.drop(columns_to_drop, axis=1)
```

```python
df['Ward/Commune'] = df['Ward/Commune'].str.replace(r"^(Xã |Phường |Thị trấn )", "", regex=True)
df['District'] = df['District'].str.replace(r"^(Quận |Huyện |Thị xã |Thành phố )", "", regex=True)
df['Province'] = df['Province'].str.replace(r"^(Tỉnh |Thành phố )", "", regex=True)
```

```python
file_name = "TinhHuyenXa.csv"
df.to_csv(file_name, index=False, encoding='utf-8-sig')
```

```python
housing_data = pd.read_csv('vietnam_housing_dataset2.csv')
location_data = pd.read_csv('TinhHuyenXa.csv')

merged_data = pd.merge(housing_data, location_data, how='left', 
                       on=['Province', 'District', 'Ward/Commune'])
```

```python
merged_data.head()
```

```python
merged_data = merged_data.dropna(subset=['Ma_TP', 'Ma_QH', 'Ma_PX', 'Province'])
```

```python
file_name = "vietnam_housing_dataset_final.csv"
merged_data.to_csv(file_name, index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv('./vietnam_housing_dataset_final.csv')
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
categorical_columns = df.select_dtypes(include=[object]).columns
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
columns_to_drop = ['Ward/Commune', 'District', 'Province']

df = df.drop(columns_to_drop, axis=1)
```

```python
file_name = "vietnam_housing_dataset_final_2.csv"
df.to_csv(file_name, index=False, encoding='utf-8-sig')
```
